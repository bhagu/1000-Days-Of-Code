```python
states_babies = pd.read_csv("state_baby_names.csv")
```

```python
states_babies.head(20)
```

```python
ca_babies_2014 = states_babies.loc[states_babies["Year"] == 2014,:].loc[states_babies["State"] == "CA", :].sort_values("Count", ascending = False)
```

```python
ca_babies_2014.iloc[0:5]
```

```python
ca_babies = states_babies.loc[states_babies["State"] == "CA", :]
```

```python
ca_babies
```

```python
def popular(s):
  return s.iloc[0]
```

```python
ca_grouped = ca_babies.sort_values("Count", ascending = False).groupby(["Year", "Gender"]).agg(popular)
```

**Visualize Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv("Airbnb_NYC_2019.csv")
```

```python
listings
```

```python
sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.hist(listings['price'])
plt.xlabel('price in usd')
plt.show()
```

```python
plt.hist(listings['price'], bins = np.arange(0, 1100, 40))
plt.xlabel('price in usd')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```
